## Section 1 — Full Asset Price Data Pull (yfinance)
Pulling complete dataset across equities, FX, commodities.

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Define tickers by asset class
equities = ['SPY', 'EEM']
commodities = ['GLD', 'CL=F']
fx = ['DX-Y.NYB', 'EURUSD=X', 'USDJPY=X', 'USDTRY=X', 'USDBRL=X']

# Combine
all_tickers = equities + commodities + fx

print("Tickers to download:", all_tickers)
print("Total series:", len(all_tickers))
```

```python
# Download all tickers from 2000 to present
raw_data = yf.download(all_tickers, start='2000-01-01')['Close']

# Rename columns for clarity
raw_data.columns = ['SPY', 'EEM', 'Gold', 'Oil_WTI', 'DXY', 'EURUSD', 'USDJPY', 'USDTRY', 'USDBRL']

# Quick check
print("Shape:", raw_data.shape)
print("\nDate range:", raw_data.index[0], "to", raw_data.index[-1])
print("\nFirst few rows:")
raw_data.head()
```

```python
# Check first valid date for each series
print("First valid date per series:")
for col in raw_data.columns:
    first_valid = raw_data[col].first_valid_index()
    print(f"  {col}: {first_valid}")
```

```python
# Document NaN counts before cleaning
print("NaN counts per series before cleaning:")
print(raw_data.isnull().sum())
print(f"\nTotal rows: {len(raw_data)}")

# Forward fill — handles weekends and bank holidays where markets are closed
# Limit of 3 days prevents filling across genuine long-term data gaps
cleaned_data = raw_data.copy()
cleaned_data = cleaned_data.ffill(limit=3)

# Check NaN counts after filling
print("\nNaN counts after forward fill:")
print(cleaned_data.isnull().sum())
```

```python
# Check if USDJPY gaps are just at the start or scattered throughout
usdjpy_nulls = cleaned_data['USDJPY'].isnull()
print("USDJPY null gaps after 2004:")
print(usdjpy_nulls['2004':].sum())
print("\nUSDBRL null gaps after 2006:")
print(cleaned_data['USDBRL']['2006':].isnull().sum())
```

```python
# Test alternative USDJPY ticker
usdjpy_alt = yf.download('JPY=X', start='2000-01-01')['Close']
print("Alternative USDJPY shape:", usdjpy_alt.shape)
print("NaN count:", usdjpy_alt.isnull().sum())
print("First valid date:", usdjpy_alt.first_valid_index())
```

```python
# Download JPY=X as clean USDJPY replacement
usdjpy_clean = yf.download('JPY=X', start='2000-01-01')['Close']
usdjpy_clean.columns = ['USDJPY']

# Replace the patchy USDJPY column in cleaned_data
cleaned_data['USDJPY'] = usdjpy_clean['USDJPY']

# Verify
print("USDJPY NaN count after swap:", cleaned_data['USDJPY'].isnull().sum())
print("Final NaN counts across all series:")
print(cleaned_data.isnull().sum())
```

```python
# Create processed directory if it doesn't exist
os.makedirs('data/processed', exist_ok=True)

# Save cleaned data as CSV
cleaned_data.to_csv('data/processed/asset_prices_clean.csv')

# Confirm save
print("Saved successfully.")
print("File location: data/processed/asset_prices_clean.csv")
print("Shape:", cleaned_data.shape)
```

```python
# Reload cleaned data
data = pd.read_csv('data/processed/asset_prices_clean.csv', index_col='Date', parse_dates=True)

# Function to plot a window around an event
def plot_event(event_date, event_name, window_days=60):
    start = pd.Timestamp(event_date) - pd.Timedelta(days=window_days)
    end = pd.Timestamp(event_date) + pd.Timedelta(days=window_days)
    
    window = data.loc[start:end, ['SPY', 'Gold', 'Oil_WTI', 'USDJPY']].copy()
    
    # Index to 100 at start of window
    window_indexed = window / window.iloc[0] * 100
    
    window_indexed.plot(figsize=(12, 5), title=f'{event_name} — ±60 days')
    plt.axvline(pd.Timestamp(event_date), color='red', linestyle='--', label='Event date')
    plt.ylabel('Index (100 = 60 days before event)')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Plot three known events
plot_event('2001-09-11', '9/11 Attacks')
plot_event('2022-02-24', 'Russia Invades Ukraine')
plot_event('2020-03-11', 'COVID Pandemic Declared')
```

```python
# Reload cleaned data
data = pd.read_csv('data/processed/asset_prices_clean.csv', index_col='Date', parse_dates=True)

def plot_event(event_date, event_name, window_days=60):
    start = pd.Timestamp(event_date) - pd.Timedelta(days=window_days)
    end = pd.Timestamp(event_date) + pd.Timedelta(days=window_days)
    
    window = data.loc[start:end, ['SPY', 'Gold', 'Oil_WTI', 'USDJPY']].copy()
    
    fig, axes = plt.subplots(2, 2, figsize=(14, 8))
    fig.suptitle(f'{event_name} — ±60 days', fontsize=14)
    
    series_axes = [('SPY', axes[0,0]), ('Gold', axes[0,1]), 
                   ('Oil_WTI', axes[1,0]), ('USDJPY', axes[1,1])]
    
    for series, ax in series_axes:
        if window[series].isnull().all():
            ax.set_title(f'{series} — no data')
            continue
        indexed = window[series] / window[series].dropna().iloc[0] * 100
        ax.plot(indexed)
        ax.axvline(pd.Timestamp(event_date), color='red', 
                   linestyle='--', label='Event date')
        ax.set_title(series)
        ax.set_ylabel('Index (100 = start)')
        ax.tick_params(axis='x', rotation=45)
    
    plt.tight_layout()
    plt.show()

plot_event('2001-09-11', '9/11 Attacks')
plot_event('2022-02-24', 'Russia Invades Ukraine')
plot_event('2020-03-11', 'COVID Pandemic Declared')
```

## Section 3 — Full Dataset Build
Complete pull of all 22 series across 6 asset classes.

```python
# ============================================================
# FULL DATASET BUILD — 22 SERIES ACROSS 6 ASSET CLASSES
# ============================================================

# --- YFINANCE PULL ---
yf_tickers = {
    'SPY':      'SPY',
    'EEM':      'EEM',
    '^STOXX50E':'EuroStoxx50',
    '^N225':    'Nikkei225',
    'GLD':      'Gold',
    'CL=F':     'Oil_WTI',
    'BZ=F':     'Oil_Brent',
    'HG=F':     'Copper',
    'DX-Y.NYB': 'DXY',
    'JPY=X':    'USDJPY',
    'USDCNY=X': 'USDCNY',
    'USDTRY=X': 'USDTRY',
    'BTC-USD':  'Bitcoin'
}

print("Downloading yfinance series...")
raw_yf = yf.download(list(yf_tickers.keys()), start='2000-01-01', progress=False)['Close']
raw_yf.columns = list(yf_tickers.values())
print(f"yfinance pull complete — shape: {raw_yf.shape}")

# --- FRED PULL ---
print("\nDownloading FRED series...")
fred_series = {
    'VIX':        fred.get_series('VIXCLS',        observation_start='2000-01-01'),
    'MOVE':       fred.get_series('BAMLMOVE',      observation_start='2000-01-01'),
    'US2YR':      fred.get_series('DGS2',          observation_start='2000-01-01'),
    'US10YR':     fred.get_series('DGS10',         observation_start='2000-01-01'),
    'US30YR':     fred.get_series('DGS30',         observation_start='2000-01-01'),
    'TIPS5YR':    fred.get_series('T5YIE',         observation_start='2000-01-01'),
    'HY_Spread':  fred.get_series('BAMLH0A0HYM2', observation_start='2000-01-01'),
    'IG_Spread':  fred.get_series('BAMLC0A0CM',   observation_start='2000-01-01'),
    'EM_Spread':  fred.get_series('BAMLEMCBPIOAS', observation_start='2000-01-01')
}

fred_data = pd.DataFrame(fred_series)
print(f"FRED pull complete — shape: {fred_data.shape}")

# --- MERGE ---
print("\nMerging datasets...")
full_data = raw_yf.join(fred_data, how='outer')

# --- CLEAN ---
full_data = full_data.ffill(limit=3)

# --- DIAGNOSTICS ---
print("\nFull dataset shape:", full_data.shape)
print("\nColumns:", list(full_data.columns))
print("\nDate range:", full_data.index[0], "to", full_data.index[-1])
print("\nNaN counts:")
print(full_data.isnull().sum())
```

```python
# --- FRED PULL ---
print("Downloading FRED series...")
fred_series = {
    'VIX':        fred.get_series('VIXCLS',        observation_start='2000-01-01'),
    'US2YR':      fred.get_series('DGS2',          observation_start='2000-01-01'),
    'US10YR':     fred.get_series('DGS10',         observation_start='2000-01-01'),
    'US30YR':     fred.get_series('DGS30',         observation_start='2000-01-01'),
    'TIPS5YR':    fred.get_series('T5YIE',         observation_start='2000-01-01'),
    'HY_Spread':  fred.get_series('BAMLH0A0HYM2', observation_start='2000-01-01'),
    'IG_Spread':  fred.get_series('BAMLC0A0CM',   observation_start='2000-01-01'),
    'EM_Spread':  fred.get_series('BAMLEMCBPIOAS', observation_start='2000-01-01')
}
fred_data = pd.DataFrame(fred_series)
print(f"FRED pull complete — shape: {fred_data.shape}")

# --- MERGE ---
print("\nMerging datasets...")
full_data = raw_yf.join(fred_data, how='outer')

# --- FILTER TO SPY TRADING DAYS ---
spy_raw = yf.download('SPY', start='2000-01-01', progress=False)['Close']
trading_days = spy_raw.dropna().index
full_data_clean = full_data.reindex(trading_days)

# --- FORWARD FILL ---
full_data_clean = full_data_clean.ffill(limit=3)

# --- DIAGNOSTICS ---
print("\nFull dataset shape:", full_data_clean.shape)
print("\nDate range:", full_data_clean.index[0], "to", full_data_clean.index[-1])
print("\nNaN counts:")
print(full_data_clean.isnull().sum())
```
