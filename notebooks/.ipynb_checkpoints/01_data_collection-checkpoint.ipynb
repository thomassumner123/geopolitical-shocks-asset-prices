# 01 — Data Collection
This notebook pulls and cleans all asset price data used in the project.
Data sources: yfinance (equities, FX, commodities) and FRED API (rates, volatility, credit spreads).
All raw data is saved to /data/raw. Cleaned, merged data is saved to /data/processed.
Last updated: March 2026

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

tickers = ['SPY', 'EEM', 'GLD', 'USO']
data = yf.download(tickers, start='2000-01-01')['Close']

print(data.head())
print(data.shape)
data.plot(figsize=(14, 6), title='Asset Prices 2000-2026')
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download each ticker individually - works better with older yfinance
SPY = yf.Ticker("SPY").history(start="2000-01-01")['Close']
EEM = yf.Ticker("EEM").history(start="2000-01-01")['Close']
GLD = yf.Ticker("GLD").history(start="2000-01-01")['Close']
USO = yf.Ticker("USO").history(start="2000-01-01")['Close']

# Combine into one dataframe
data = pd.DataFrame({
    'SPY': SPY,
    'EEM': EEM,
    'GLD': GLD,
    'USO': USO
})

print(data.head())
print(data.shape)

data.plot(figsize=(14, 6), title='Asset Prices 2000-2026')
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download data with CL=F replacing USO
tickers = ['SPY', 'EEM', 'GLD', 'CL=F']
data = yf.download(tickers, start='2000-01-01')['Close']

# Rename columns for clarity
data.columns = ['Oil_CL', 'EEM', 'GLD', 'SPY']

# Normalise each series to 100 at its first non-NaN value
data_indexed = data.copy()
for col in data_indexed.columns:
    first_valid = data_indexed[col].first_valid_index()
    data_indexed[col] = (data_indexed[col] / data_indexed[col].loc[first_valid]) * 100

# Plot
data_indexed.plot(figsize=(14, 6), title='Asset Prices Indexed to 100 (Base = First Available Date)')
plt.ylabel('Index (100 = first available date)')
plt.xlabel('Date')
plt.tight_layout()
plt.show()
```
